| Operator | Meaning                 | Example                  |
| -------- | ----------------------- | ------------------------ |
| `+=`     | add and assign          | `x += 5` → `x = x + 5`   |
| `-=`     | subtract and assign     | `x -= 2` → `x = x - 2`   |
| `*=`     | multiply and assign     | `x *= 3` → `x = x * 3`   |
| `/=`     | divide and assign       | `x /= 2` → `x = x / 2`   |
| `//=`    | floor divide and assign | `x //= 2` → `x = x // 2` |
| `%=`     | modulo and assign       | `x %= 3` → `x = x % 3`   |
| `**=`    | exponent and assign     | `x **= 2` → `x = x ** 2` |

__`join()`__ is a string method that combines(concatenates) an iterable of strings into one single string, placing a separator between each element.

````python
words = ["hello", "world"]
result = " ".join(words)
print(result)
````

````python
", ".join(["a", "b", "c"])      # "a, b, c"
"-".join(["2026", "01", "12"])  # "2026-01-12"
"".join(["p", "y", "t", "h", "o", "n"])  # "python"
"\n".join(["line1", "line2"])   # multiline string
````

````python
def list_home_town_items():
    home_town = {"city": "Miami", "state": "Florida", "population": "500000"}
    list_home_town_items = []

    for key, value in home_town.items():
        list_home_town_items.append(f"{key} = {value}")

    return list_home_town_items


````
