Here is how we declare a variable in Python:

num = 15


__Notice the lack of let or const keywords__

Variables are case sensitive 

In Python, we use snake_case

my_number = 10


We can freely assign a new value to a variable after it has been declared:

```python
my_number = 15
print(my_number)
# prints: 15
my_number = -4
print(my_number)
# prints: -4
```

 Python doesn’t have a way to declare constants, the convention is to declare any variable you’d like to be known as a constant in SCREAMING_SNAKE_CASE. does not prevent reassignment. It only communicates that the variable is not intended to be reassigned.
