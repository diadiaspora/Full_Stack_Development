__Exercise 4__

````python
def convert_temperature(temp, unit):
    if unit == "C":
        return (temp * 9/5) + 32
    else: return (temp - 32) * 5/9
        
    
````

````python
convert_temperature(0, 'C')
````

````python
convert_temperature(32, 'F')
````

Lambda version:

lambda x: value_if_true if condition else value_if_false


````python
convert_temperature = lambda temp, unit : (temp * 9/5) + 32 if unit == "C" else (temp - 32) * 5/9
````

````python
convert_temperature(12, 'C') 
````

````python
def basic_calculator(a, b, operation):
    operations = {
        "add": a + b,
        "subtract": a - b,
        "multiply": a * b,
        "divide": a / b
    }

    return operations[operation]
````

````python
basic_calculator(10, 5, 'subtract')
````

```python
}
```

3. Use the argument operation to select a key in the dictionary:

4. Return the result of the lookup:



* Python finds the key in the dictionary

* Returns the corresponding value

5. Example of a function call:

```python
basic_calculator(10, 5, "multiply")
```


* operation argument = "multiply"

* Dictionary operations = {"add": 15, "subtract": 5, "multiply": 50, "divide": 2.0}

* operations["multiply"] â†’ 50 â†’ returned by the function

````python
import operator

def basic_calculator(a, b, operation):
    operations = {
        "add": operator.add,
        "subtract": operator.sub,
        "multiply": operator.mul,
        "divide": operator.truediv
    }

    return operations[operation](a, b)
````
