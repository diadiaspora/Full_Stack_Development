__Exercise 1__

    

print only version:

```python
def calculate_area_triangle(base, height):
    print((base * height) / 2)
    
```

```python
calculate_area_triangle(10, 5) 
```

```python
calculate_area_triangle(7, 3) 
```

Correct (with return)

```python
def calculate_area_triangle(base, height):
    return (base * height) / 2
```

```python
calculate_area_triangle(12, 8) 
```

Lambda version:

```python
calculate_area_triangle = lambda base, height: (base * height) / 2
```

```python
area = calculate_area_triangle(10, 8)
print(area)
```

__Exercise 2__

```python
def simple_interest(princ, rate, time):
    return princ * rate * time / 100.
```

```python
simple_interest(1000, 5, 2)
```

Lambda version

```python
simple_interest = lambda princ, rate, time: princ * rate * time / 100
```

```python
simple_interest(1300, 3, 3)
```

__Exercise 3__

```python
def apply_discount(price, discount_percentage):
    discount = price * discount_percentage / 100
    return price - discount
    
```

```python
apply_discount(100, 25)
```

Lambda version

```python
apply_discount = lambda price, discount_percentage: price - (price * discount_percentage / 100)
```

```python
apply_discount(80, 10)
```

__Exercise 4__

```python
def convert_temperature(temp, unit):
    if unit == "C":
        return (temp * 9/5) + 32
    else: return (temp - 32) * 5/9
        
    
```

```python
convert_temperature(0, 'C')
```

```python
convert_temperature(32, 'F')
```

Lambda version:

lambda x: value_if_true if condition else value_if_false


```python
convert_temperature = lambda temp, unit : (temp * 9/5) + 32 if unit == "C" else (temp - 32) * 5/9
```

```python
convert_temperature(12, 'C') 
```
