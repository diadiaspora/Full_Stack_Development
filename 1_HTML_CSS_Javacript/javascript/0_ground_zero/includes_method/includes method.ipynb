The .includes() method checks if a value exists inside a string or an array.

Use .includes() when the question is basically:

“Does this exist inside this?”

If you see phrases like:

“contains”
“includes”
“has”
“is inside”
“exists in”

It returns:

true → if the value is found

false → if it is not

'hello'.includes('e')   // true

'hello'.includes('x')   // false

'javascript'.includes('script') // true

[1, 2, 3].includes(2) // true

[1, 2, 3].includes(5) // false

['a', 'e', 'i'].includes('e') // true

“.includes() and hash maps solve the same problem (existence check),
but .includes() uses brute force, while hash maps are optimized.”
